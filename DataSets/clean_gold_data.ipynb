```python
# Clean Gold Data Function.
def clean_data(df):
    df["month"]=df.index.month
    df["day"]=df.index.day
    df["year"]=df.index.year
    df_grp=df.groupby(["year","month","day"]).first()
    df_grp=df_grp.dropna()
    df_grp['Gold Price Fix AM in US dollars per troy ounce'] = df_grp['Gold Price Fix AM in US dollars per troy ounce'].str.replace(r"[a-zA-Z]",'')
    df_grp['Gold Price Fix AM in US dollars per troy ounce'] = df_grp['Gold Price Fix AM in US dollars per troy ounce'].str.replace("$",'')
    df_grp['Gold Price Fix PM in US dollars per troy ounce'] = df_grp['Gold Price Fix PM in US dollars per troy ounce'].str.replace(r"[a-zA-Z]",'')
    df_grp['Gold Price Fix PM in US dollars per troy ounce'] = df_grp['Gold Price Fix PM in US dollars per troy ounce'].str.replace("$",'')
    df_grp['Gold Price Fix PM in US dollars per troy ounce'] = df_grp['Gold Price Fix PM in US dollars per troy ounce'].str.replace(",",'')
    df_grp['Gold Price Fix AM in US dollars per troy ounce'] = df_grp['Gold Price Fix AM in US dollars per troy ounce'].str.replace(",",'')
    df_grp['Gold Price Fix AM in US dollars per troy ounce'] = df_grp['Gold Price Fix AM in US dollars per troy ounce'].astype(float)
    df_grp['Gold Price Fix PM in US dollars per troy ounce'] = df_grp['Gold Price Fix PM in US dollars per troy ounce'].astype(float)
    return df_grp.to_csv(f"{df_grp.index[0][0]}_Gold_Monthly")   
```

```python
# Clean Gold Data Function.
def clean_data_18(df):
    df["month"]=df.index.month
    df["day"]=df.index.day
    df["year"]=df.index.year
    df_grp=df.groupby(["year","month","day"]).first()
    df_grp=df_grp.dropna()
    df_grp['Daily Gold Price LBMA AM FIX'] = df_grp['Daily Gold Price LBMA AM FIX'].str.replace(r"[a-zA-Z]",'')
    df_grp['Daily Gold Price LBMA AM FIX'] = df_grp['Daily Gold Price LBMA AM FIX'].str.replace("$",'')
    df_grp['Daily Gold Price LBMA PM FIX'] = df_grp['Daily Gold Price LBMA PM FIX'].str.replace(r"[a-zA-Z]",'')
    df_grp['Daily Gold Price LBMA PM FIX'] = df_grp['Daily Gold Price LBMA PM FIX'].str.replace("$",'')
    df_grp['Daily Gold Price LBMA PM FIX'] = df_grp['Daily Gold Price LBMA PM FIX'].str.replace(",",'')
    df_grp['Daily Gold Price LBMA AM FIX'] = df_grp['Daily Gold Price LBMA AM FIX'].str.replace(",",'')
    df_grp['Daily Gold Price LBMA PM FIX'] = df_grp['Daily Gold Price LBMA PM FIX'].str.replace("--",'0')
    df_grp['Daily Gold Price LBMA AM FIX'] = df_grp['Daily Gold Price LBMA AM FIX'].str.replace("--",'0')
    df_grp['Daily Gold Price LBMA AM FIX'] = df_grp['Daily Gold Price LBMA AM FIX'].astype(float)
    df_grp['Daily Gold Price LBMA PM FIX'] = df_grp['Daily Gold Price LBMA PM FIX'].astype(float)
    return df_grp.to_csv(f"{df_grp.index[0][0]}_Gold_Monthly")   

```
